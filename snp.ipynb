```python
ref_alt = new_df.iloc[:,:5]

class_list = []

for r,a in zip(ref_alt['REF'],ref_alt['ALT']):
    # T>G:
    if (r == 'C' and a == 'T') or (r == 'G' and a == 'A'):
        class_list.append('C>T')
    elif (r == 'C' and a == 'A') or (r == 'G' and a == 'T'):
        class_list.append('C>A')
    elif (r == 'C' and a == 'G') or (r == 'G' and a == 'C'):
        class_list.append('C>G')
    elif (r == 'T' and a == 'A') or (r == 'A' and a == 'T'):
        class_list.append('T>A')
    elif (r == 'T' and a == 'C') or (r == 'A' and a == 'G'):
        class_list.append('T>C')
    elif (r == 'T' and a == 'G') or (r == 'A' and a == 'C'):
        class_list.append('T>G')
    elif len(r) < len(a):
        class_list.append('Insertion')
#         print('insertion')
    elif len(r) > len(a):
        class_list.append('Deletion')
        pass
    else:
        class_list.append('other')

ref_alt['Class'] = class_list
ref_alt
```
