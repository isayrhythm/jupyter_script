```python

rnk_file = r'/home/luotao/stock/data/tim-n2.csv'
# gsea_data.gsea_data.rnk 是已经排好序的数据
rnk = np.loadtxt(rnk_file,str,delimiter = ',',skiprows=1)
new_rnk = np.empty((2,))
for i in rnk:
    if i[0] == '':
        pass
    else:
        new_rnk = np.vstack((new_rnk,np.mat([i[0],float(i[1])])))
len(new_rnk)
new_rnk = new_rnk[1:]
new_rnk = new_rnk.astype('object')  # <U32类型没办法转浮点，坑了我一上午

# 将第二列字符串转换为浮点型

new_rnk[:,1] = new_rnk[:, 1].astype(float)
new_rnk
```
