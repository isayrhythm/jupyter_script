```python
import pydeseq2,os
import pandas as pd
```

```python
count_path = '/home/luotao/c_elegans/RNA-seq/SRA_yy/count_out/'
count_list = os.listdir(count_path)

count_list.sort()

count_df = pd.read_table(count_path+sm,skipfooter=5)

matrix = pd.DataFrame()
for sm in count_list:
    count_df = pd.read_table(count_path+sm,skipfooter=5)
    count_df.columns = ['gene_id',sm.replace('.count','')]
    if matrix.empty:
        matrix = count_df
    else:
        matrix = pd.merge(matrix, count_df , on='gene_id')
matrix
```

```python
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats

df_T = matrix.transpose()

col_slice = matrix.iloc[:,1:]
row_sum = col_slice.sum(axis=1)

# 使用布尔索引选择行总和大于等于 10 的行
new_df = matrix[row_sum >= 10]

#转置
new_df = new_df.transpose()

# 设置行名
new_df.columns = list(new_df.iloc[0])
# 取前6个样品数据，删除第一行
new_df = new_df.iloc[1:]

new_df
```

```python



condition_df = pd.DataFrame({'condition': ['A', 'A', 'A','C','C','C','B','B','B']})

condition_df.index = new_df.index
condition_df
```

```python
dds = DeseqDataSet(counts = new_df, clinical=condition_df, design_factors="condition")
```

```python
dds.deseq2()
```

```python
stat_res = DeseqStats(dds, n_cpus=8)

df_summary = stat_res.summary()
```
